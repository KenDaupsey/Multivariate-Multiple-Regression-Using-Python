*** Multivariate Multiple Regression ***

```python
### Import Necessary Libraries

import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
### Load Dataset and Examine First 5 Rows
url= "https://raw.githubusercontent.com/KenDaupsey/Multivariate-Multiple-Regression-Using-Python/main/hsb2%7Edata.csv"

df = pd.read_csv(url)
df.head()
```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming the dataset is already loaded into a pandas DataFrame called 'df'

# Perform one-hot encoding for the 'female' variable
df = pd.get_dummies(df, columns=['female'], drop_first=True)

# Define the independent variables, including the intercept
X = sm.add_constant(df[['female_male', 'math', 'science', 'socst']])

# Define the dependent variables
Y_write = df['write']
Y_read = df['read']

# Fit the multivariate multiple regression model for 'write'
model_write = sm.OLS(Y_write, X).fit()

# Fit the multivariate multiple regression model for 'read'
model_read = sm.OLS(Y_read, X).fit()

# Print the summary results
print('Summary for the "write" model:')
print(model_write.summary())

print('\nSummary for the "read" model:')
print(model_read.summary())
```
